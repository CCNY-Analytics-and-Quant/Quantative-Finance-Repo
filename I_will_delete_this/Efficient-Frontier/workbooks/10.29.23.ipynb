```python
# packages
import pandas as pd
import yfinance as yf
import numpy as np

import seaborn as sn
import matplotlib.pyplot as plt
```

```python
# Step 1: Choose seven (7) assets
tickers = ["NVDA", "NFLX", "GME", "BB", "NOK", "AMZN", "TSLA"]
tickers
```

```python
# Step 2: Retrieve daily open or close data on your assets for the previous 2 years
df = yf.download(tickers, period='2y')

adj_close = df['Adj Close']
returns = adj_close.pct_change()
returns
```

```python
returns = np.log(1 + adj_close.pct_change())
returns
```

```python
# Step 3: Calculate the mean, variance, and correlation matrix for all assets
table = pd.DataFrame(index=tickers)

table['Mean'] = returns.mean()
table['Variance'] = returns.var()
```

```python
# Displays table result - mean and variance 
table
```

```python
# Displays table result - correlation
corr_matrix = returns.corr()
corr_matrix
```

```python
plt.figure()
sn.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
